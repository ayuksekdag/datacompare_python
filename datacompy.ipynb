```python
df1 = pd.DataFrame ({'loanid': [2, 4, 8, 0],
                   'loanamt': [11, 10, 12, 14]},
                  )

df2 = pd.DataFrame ({'loanid': [2, 4, 8, 0],
                   'loanamt': [11, 10, 12, 13]},
                  )
```

```python
df1
```

```python
df2
```

```python
import datacompy

compare = datacompy.Compare(
  df1,
  df2,
  join_columns='loanid', #You can also specify a list of columns
  abs_tol=0.0001,
  rel_tol=0,
  df1_name='original',
  df2_name='new'
)

print(compare.report())
```
